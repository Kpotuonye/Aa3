```python
import folium
```

```python
# To create a base map, simply pass your coordinates to Folium.
# passing the Latitude & Longitude of Salzburg City (47.811195° N, 13.033229° E) and create a Map.

folium.Map(location=[47.811195, 13.033229])
```

```python
folium.Map(location=[47.811195, 13.033229], tiles= 'OpenStreetMap')
```

```python
folium.Map(location=[47.811195, 13.033229], tiles='Stamen Terrain')
```

```python
# To set the default Zoom level to 15

salzburg_map = folium.Map(location=[47.811195, 13.033229], tiles="Openstreetmap", zoom_start=15, control_scale=True)

salzburg_map
```

```python
# Marking My area of interest within Salzburg city

#Randomly chosen populated areas

# Mirabellgarden
folium.Marker(location=[47.8054, 13.0417],
    popup="Mirabellgarden",
    icon=folium.Icon(color="blue", icon="flag"),
).add_to(salzburg_map)

#Europark
folium.Marker(location=[47.8154, 13.0052],
    popup="Europark",
    icon=folium.Icon(color="green", icon="flag"),
).add_to(salzburg_map)

#Hauptbahnhof
folium.Marker(location=[47.8131, 13.0451],
    popup="Hauptbahnhof",
    icon=folium.Icon(color="red", icon="flag"),
).add_to(salzburg_map)

#CityBeatsClub
folium.Marker(location=[47.8007, 13.0413],
    popup="CityBeatClub",
    icon=folium.Icon(color="pink", icon="flag"),
).add_to(salzburg_map)

salzburg_map
# Displaying the Map with the locations
```
